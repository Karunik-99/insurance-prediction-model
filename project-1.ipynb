```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=pd.read_csv("/content/insurance.csv")
```

```python
data.head()
```

```python
data['sex']=data['sex'].apply({'female':1,'male':0}.get)
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
data['region']=data['region'].apply({'southwest':0,'southeast':1,'northwest':2,'northeast':3}.get)
```

```python
data.head(40)
```

```python
x=data[['age','sex','bmi','children','smoker','region']] #independent
y=data['expenses'] #dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.78)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
y_pred=regression.predict(x_test)
```

```python
print(y_pred)
```
